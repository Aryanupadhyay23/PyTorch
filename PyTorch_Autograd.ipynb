# What is Autograd?

Autograd is PyTorch’s automatic differentiation engine that automatically computes gradients for tensor operations.

In deep learning, neural networks learn by minimizing errors. To reduce these errors, the model needs gradients, which represent how much a parameter affects the final output. Computing these gradients manually for complex neural networks is extremely difficult. Autograd solves this problem by automatically calculating derivatives during backpropagation.

PyTorch tracks all operations performed on tensors that have:

```python
requires_grad=True
```

Using these operations, PyTorch creates a computational graph dynamically during execution. When the `backward()` function is called, Autograd traverses this graph in reverse order and computes gradients using the chain rule.

---

# Features of Autograd

- Automatic gradient computation
- Dynamic computational graph creation
- Supports backpropagation
- Simplifies neural network training
- Eliminates manual derivative calculations
- Works efficiently with complex tensor operations

---

# Uses of Autograd

## 1. Gradient Computation

Autograd automatically computes derivatives of tensors with respect to model parameters.

## 2. Backpropagation

It is used during backpropagation to calculate gradients for updating weights and biases.

## 3. Neural Network Training

Autograd helps train deep learning models using optimization algorithms such as Gradient Descent and Adam.

## 4. Optimization of Parameters

The computed gradients are used by optimizers to minimize the loss function.

## 5. Handling Complex Mathematical Operations

It can efficiently compute gradients for deeply nested mathematical operations and large neural networks.

---

# Simple Example of Autograd

```python
import torch

# Create tensor with gradient tracking enabled
x = torch.tensor(2.0, requires_grad=True)

# Perform operation
y = x ** 2

# Compute gradients
y.backward()

# Print derivative dy/dx
print(x.grad)
```

Output:

```python
tensor(4.)
```

Explanation:

\[
y = x^2
\]

\[
\frac{dy}{dx} = 2x
\]

For:

\[
x = 2
\]

\[
\frac{dy}{dx} = 4
\]

Autograd automatically computes this derivative without manually writing differentiation logic.

````python
import torch

# Inputs
x = torch.tensor(6.7)   # Input feature
y = torch.tensor(0.0)   # True label (binary)

w = torch.tensor(1.0)   # Weight
b = torch.tensor(0.0)   # Bias
````

````python
# Binary Cross-Entropy Loss for scalar
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-8  # To prevent log(0)
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return -(target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))
````

````python
# Forward pass
z = w * x + b   # Weighted sum (linear part)
y_pred = torch.sigmoid(z)   # Predicted probability

# Compute binary cross-entropy loss
loss = binary_cross_entropy_loss(y_pred, y)
````

````python
loss
````

````python
# Derivatives:

# 1. dL/d(y_pred): Loss with respect to the prediction (y_pred)
dloss_dy_pred = (y_pred - y) / (y_pred * (1 - y_pred))

# 2. dy_pred/dz: Prediction (y_pred) with respect to z (sigmoid derivative)
dy_pred_dz = y_pred * (1 - y_pred)

# 3. dz/dw and dz/db: z with respect to w and b
dz_dw = x   # dz/dw = x
dz_db = 1   # dz/db = 1 (bias contributes directly to z)

# Final gradients using chain rule
dL_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dL_db = dloss_dy_pred * dy_pred_dz * dz_db
````

````python
print(f"Manual Gradient of loss w.r.t weight (dw): {dL_dw}")
print(f"Manual Gradient of loss w.r.t bias (db): {dL_db}")
````
