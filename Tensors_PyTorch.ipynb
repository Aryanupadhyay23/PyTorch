# What are Tensors ?

A tensor is a specialised multi-dimensional array designed for mathematical and computational efficiency. In deep learning and machine learning, tensors are used to store and process data efficiently. The dimension of a tensor represents in how many directions the data is spanning.

For example:
- A scalar has 0 dimensions.
- A vector has 1 dimension.
- A matrix has 2 dimensions.

As dimensions increase, tensors can represent more complex data like images, batches of images, and videos.

## Real World Examples of Tensors

### 1. Scalars : 0-Dimensional Tensors

A scalar is a single numerical value. It does not have any direction or axis.

- Represents a single value, often used for simple metrics or constants.
- Commonly used for loss values, accuracy scores, or constants.

**Example:**
- Loss value after a forward pass in a neural network.
- Values like `5.0` or `-3.14`

---

### 2. Vectors : 1-Dimensional Tensors

A vector is a collection of numbers arranged in a single direction.

- Represents a sequence or collection of values.
- Frequently used in embeddings and feature representations.

**Example:**
- Word embeddings in NLP.

```python
[0.12, -0.84, 0.33]
```

This vector can represent the semantic meaning of a word using models like Word2Vec or GloVe.

---

### 3. Matrices : 2-Dimensional Tensors

A matrix is a grid-like structure containing rows and columns.

- Represents tabular or structured data.
- Often used to represent grayscale images.

**Example:**

```python
[
 [0, 255, 128],
 [34, 90, 180]
]
```

Each value represents the intensity of a pixel in a grayscale image.

---

### 4. 3D Tensors : Coloured Images

A 3D tensor adds another dimension to store channels.

- Commonly used for RGB images.
- Shape format:

```python
[height, width, channels]
```

**Example:**

```python
[256, 256, 3]
```

Here:
- `256 × 256` = image size
- `3` = RGB colour channels

---

### 5. 4D Tensors : Batches of Images

A 4D tensor adds the batch dimension to 3D image data.

- Used when training neural networks on multiple images together.
- Shape format:

```python
[batch_size, height, width, channels]
```

**Example:**

```python
[32, 128, 128, 3]
```

This means:
- 32 images
- Each image size is `128 × 128`
- 3 RGB channels

---

### 6. 5D Tensors : Video Data

A 5D tensor adds a time dimension to represent sequential frames.

- Used in video processing and action recognition tasks.
- Shape format:

```python
[batch_size, frames, height, width, channels]
```

**Example:**

```python
[10, 16, 64, 64, 3]
```

This represents:
- 10 video clips
- Each clip contains 16 frames
- Each frame size is `64 × 64`
- 3 RGB channels

# Why are Tensors Useful ?

Tensors are the foundation of deep learning because they allow efficient storage and computation of data.

## Mathematical Operations

Tensors support efficient mathematical operations such as:
- Addition
- Multiplication
- Dot products
- Broadcasting
- Matrix multiplication

These operations are essential for neural network computations.

---

## Representation of Real World Data

Different types of real-world data can be represented using tensors:

- Images → 3D tensors
- Text → 2D or 3D tensors
- Audio → Sequential tensors
- Videos → 5D tensors

This makes tensors highly flexible for AI applications.

---

## Efficient Computation

Tensor operations are highly optimized for hardware acceleration.

- GPUs and TPUs process tensors very efficiently.
- This significantly speeds up deep learning training and inference.

# Where are Tensors Used in Deep Learning ?

## Data Storage

Training data such as:
- Images
- Text
- Audio
- Videos

are stored in tensor format.

---

## Weights and Biases

The learnable parameters of neural networks, such as:
- Weights
- Biases

are stored as tensors.

---

## Matrix Operations

Neural networks perform operations like:
- Matrix multiplication
- Dot products
- Broadcasting

All these computations are performed using tensors.

---

## Training Process

During training:
- Tensors flow through the network during forward propagation.
- Gradients are computed as tensors during backpropagation.

These gradients help the model learn and update its parameters.

## Tensor Data Types

| Data Type | Dtype | Description |
|---|---|---|
| 32-bit Floating Point | `torch.float32` | Standard floating-point type used for most deep learning tasks. Provides a balance between precision and memory usage. |
| 64-bit Floating Point | `torch.float64` | Double-precision floating point. Useful for high-precision numerical tasks but uses more memory. |
| 16-bit Floating Point | `torch.float16` | Half precision floating point. Commonly used in mixed-precision training to reduce memory and computational overhead on modern GPUs. |
| BFloat16 | `torch.bfloat16` | Brain floating-point format with reduced precision compared to `float16`. Used in mixed-precision training, especially on TPUs. |
| 8-bit Floating Point | `torch.float8` | Ultra-low-precision floating point. Used for experimental applications and extreme memory-constrained environments. |
| 8-bit Integer | `torch.int8` | 8-bit signed integer. Used for quantized models to save memory and computation in inference. |
| 16-bit Integer | `torch.int16` | 16-bit signed integer. Useful for special numerical tasks requiring intermediate precision. |
| 32-bit Integer | `torch.int32` | Standard signed integer type. Commonly used for indexing and general-purpose numerical tasks. |
| 64-bit Integer | `torch.int64` | Long integer type. Often used for large indexing arrays or tasks involving large numbers. |
| 8-bit Unsigned Integer | `torch.uint8` | 8-bit unsigned integer. Commonly used for image data (e.g., pixel values between 0 and 255). |
| Boolean | `torch.bool` | Boolean type, stores `True` or `False` values. Often used for masks in logical operations. |
| Complex 64 | `torch.complex64` | Complex number type with 32-bit real and 32-bit imaginary parts. Used for scientific and signal processing tasks. |
| Complex 128 | `torch.complex128` | Complex number type with 64-bit real and 64-bit imaginary parts. Offers higher precision but uses more memory. |
| Quantized Integer | `torch.qint8` | Quantized signed 8-bit integer. Used in quantized models for efficient inference. |
| Quantized Unsigned Integer | `torch.quint8` | Quantized unsigned 8-bit integer. Often used for quantized tensors in image-related tasks. |

````python
import torch
import time

# Define the size of the matrices
size = 10000  # Large size for performance comparison

# Create random matrices on CPU
matrix_cpu1 = torch.randn(size, size)
matrix_cpu2 = torch.randn(size, size)

# Measure time on CPU
start_time = time.time()
result_cpu = torch.matmul(matrix_cpu1, matrix_cpu2)  # Matrix multiplication on CPU
cpu_time = time.time() - start_time

print(f"Time on CPU: {cpu_time:.4f} seconds")

# Move matrices to GPU
matrix_gpu1 = matrix_cpu1.to('cuda')
matrix_gpu2 = matrix_cpu2.to('cuda')

# Measure time on GPU
start_time = time.time()
result_gpu = torch.matmul(matrix_gpu1, matrix_gpu2)  # Matrix multiplication on GPU
torch.cuda.synchronize()  # Ensure all GPU operations are complete
gpu_time = time.time() - start_time

print(f"Time on GPU: {gpu_time:.4f} seconds")

# Compare results
print("\nSpeedup (CPU time / GPU time):", cpu_time / gpu_time)
````
